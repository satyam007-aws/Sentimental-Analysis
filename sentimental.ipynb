```python
from textblob import TextBlob
```

```python
from newspaper import Article
```

```python
import nltk
nltk.download('punkt')
```

```python
url="https://economictimes.indiatimes.com/news/economy/policy/fighting-inflation-india-looks-to-rains-not-just-rates/articleshow/92221553.cms"
article=Article(url)
article.download()
article.parse()
article.nlp()
text =article.text
print(text)
```

```python
blob=TextBlob(text)
sentiment=blob.sentiment # -1 to 1
print(sentiment)
```
